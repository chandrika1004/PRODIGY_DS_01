#TASK: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population

Importing the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('world_population.csv')
data
```

##Visualizing a bar graph of Top 10 Countries by Population in 2022

```python
population_data=data[['Country/Territory','2022 Population']]
population_data=population_data.sort_values(by='2022 Population',ascending=False)
top_10=population_data.head(10)
plt.figure(figsize=(5,5))
sns.barplot(x='2022 Population',y='Country/Territory',data=top_10,color='green')
plt.xlabel('Population in 2022')
plt.ylabel('Country')
plt.title('Top 10 countries by population in 2022')
plt.show()
```

##Visualizing a bar graph of Population of India Over Years

```python
country_data=data[data['Country/Territory']=='India']
years=['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']
population=country_data[years].values.flatten()
plt.bar(years,population,color=['violet','pink','blue','green','yellow','orange','red','black'])
plt.xlabel('Year')
plt.ylabel('Population of India')
plt.title('Population of India Over Years')
plt.xticks(rotation=90)
plt.show()
```

##Visualizing a bar graph of Top 10 Countries by Area

```python
area_data=data[['Country/Territory','Area (km²)']]
area_data=area_data.sort_values(by='Area (km²)',ascending=False)
top_10=area_data.head(10)
plt.figure(figsize=(5,5))
sns.barplot(x='Area (km²)',y='Country/Territory',data=top_10,color='orange')
plt.xlabel('Area of countries by Area (km²)')
plt.ylabel('Country')
plt.title('Top 10 countries by Area')
plt.show()
```
